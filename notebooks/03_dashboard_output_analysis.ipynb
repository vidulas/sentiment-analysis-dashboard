# Dashboard Output Analysis

This notebook reviews the dashboard-ready outputs created in Stage 4. These files are easier to connect to dashboard tools because the main calculations are already prepared.

## 1. Why Dashboard-Ready Outputs Are Created

Raw review-level data is useful, but dashboards also need summary tables for charts, KPIs, and product comparisons. Creating separate output files keeps the dashboard simpler and makes the project easier to explain.

## 2. Import Libraries

Use pandas to read the dashboard outputs, matplotlib to create simple charts, and json to load KPI values.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

## 3. Load Dashboard Outputs

Load the dashboard-ready CSV and JSON files created by `src/create_dashboard_outputs.py`.

```python
output_dir = "../data/processed/dashboard_outputs"

dashboard_main = pd.read_csv(f"{output_dir}/dashboard_main.csv")
sentiment_distribution = pd.read_csv(f"{output_dir}/sentiment_distribution.csv")
rating_distribution = pd.read_csv(f"{output_dir}/rating_distribution.csv")
product_summary = pd.read_csv(f"{output_dir}/product_sentiment_summary.csv")
top_positive_reviews = pd.read_csv(f"{output_dir}/top_positive_reviews.csv")
top_negative_reviews = pd.read_csv(f"{output_dir}/top_negative_reviews.csv")
sentiment_match_summary = pd.read_csv(f"{output_dir}/sentiment_match_summary.csv")

with open(f"{output_dir}/dashboard_kpis.json", "r", encoding="utf-8") as file:
    dashboard_kpis = json.load(file)
```

## 4. Overall KPIs

These KPIs summarize the full review dataset and are useful for dashboard cards.

```python
dashboard_kpis
```

## 5. Sentiment Distribution

This chart shows how many reviews VADER classified as positive, neutral, or negative.

```python
sentiment_distribution.plot(
    kind="bar",
    x="vader_sentiment",
    y="count",
    figsize=(8, 4),
    color="steelblue",
    legend=False,
)
plt.title("VADER Sentiment Distribution")
plt.xlabel("VADER Sentiment")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=0)
plt.show()

sentiment_distribution
```

## 6. Rating Distribution

This chart shows how review ratings are distributed from 1 to 5.

```python
rating_distribution.plot(
    kind="bar",
    x="rating",
    y="count",
    figsize=(8, 4),
    color="seagreen",
    legend=False,
)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=0)
plt.show()

rating_distribution
```

## 7. Top Products by Review Count

These are the most reviewed products among products with at least 10 reviews.

```python
top_reviewed_products = product_summary.sort_values("review_count", ascending=False).head(10)

top_reviewed_products.sort_values("review_count").plot(
    kind="barh",
    x="product_name",
    y="review_count",
    figsize=(10, 6),
    color="darkorange",
    legend=False,
)
plt.title("Top 10 Products by Review Count")
plt.xlabel("Review Count")
plt.ylabel("Product Name")
plt.show()

top_reviewed_products
```

## 8. Top Products by Negative Percentage

These products have the highest percentage of VADER-negative reviews among products with at least 10 reviews.

```python
top_negative_products = product_summary.sort_values(
    ["negative_percentage", "review_count"], ascending=[False, False]
).head(10)

top_negative_products.sort_values("negative_percentage").plot(
    kind="barh",
    x="product_name",
    y="negative_percentage",
    figsize=(10, 6),
    color="firebrick",
    legend=False,
)
plt.title("Top 10 Products by Negative Review Percentage")
plt.xlabel("Negative Percentage")
plt.ylabel("Product Name")
plt.show()

top_negative_products
```
