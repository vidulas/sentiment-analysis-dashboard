# Data Inspection

This notebook performs a first inspection of the raw sentiment dataset. The goal is to understand the data structure before cleaning, sentiment analysis, modeling, or dashboard work.

## 1. Import Libraries

Import the main libraries needed for basic data inspection and simple visual checks.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Load Dataset

Load the raw dataset from the `data/raw` folder.

```python
data_path = "../data/raw/sentiment.csv"
df = pd.read_csv(data_path)
```

Display the first 5 rows to get a quick look at the dataset.

Check the number of rows and columns in the dataset.

List all column names so we can understand what fields are available.

Review the data type of each column.

```python
df.dtypes
```

## 7. Missing Values

Check how many missing values are present in each column.

```python
df.isnull().sum().sort_values(ascending=False)
```

## 8. Sentiment Value Counts

Count how many reviews belong to each sentiment category, if the sentiment column exists.

```python
sentiment_columns = [col for col in df.columns if col.lower() == "sentiment"]

if sentiment_columns:
    sentiment_col = sentiment_columns[0]
    print(f"Using sentiment column: {sentiment_col}")
    display(df[sentiment_col].value_counts(dropna=False))
else:
    print("No column named 'Sentiment' was found.")
```

## 9. Sentiment Percentage Distribution

Show the percentage distribution of sentiment categories, if the sentiment column exists.

```python
if sentiment_columns:
    sentiment_percentages = df[sentiment_col].value_counts(normalize=True, dropna=False) * 100
    display(sentiment_percentages.round(2))

    sentiment_percentages.plot(kind="bar", figsize=(8, 4), color="steelblue")
    plt.title("Sentiment Percentage Distribution")
    plt.xlabel("Sentiment")
    plt.ylabel("Percentage")
    plt.xticks(rotation=0)
    plt.show()
else:
    print("No column named 'Sentiment' was found.")
```

## 10. Rating Distribution

Show the rating distribution if the `Rate` column exists.

```python
if "Rate" in df.columns:
    display(df["Rate"].value_counts(dropna=False).sort_index())

    df["Rate"].value_counts(dropna=False).sort_index().plot(kind="bar", figsize=(8, 4), color="seagreen")
    plt.title("Rating Distribution")
    plt.xlabel("Rating")
    plt.ylabel("Number of Reviews")
    plt.xticks(rotation=0)
    plt.show()
else:
    print("The 'Rate' column was not found.")
```

## 11. Top 10 Most Reviewed Products

Show the 10 products with the most reviews if the `ProductName` column exists.

```python
if "ProductName" in df.columns:
    top_products = df["ProductName"].value_counts().head(10)
    display(top_products)

    top_products.sort_values().plot(kind="barh", figsize=(10, 6), color="darkorange")
    plt.title("Top 10 Most Reviewed Products")
    plt.xlabel("Number of Reviews")
    plt.ylabel("Product Name")
    plt.show()
else:
    print("The 'ProductName' column was not found.")
```
